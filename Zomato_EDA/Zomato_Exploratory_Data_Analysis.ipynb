```python
df_country = pd.read_excel('Country-Code.xlsx')
df_country
```

Here in two table country_code is same column so we combine two tables using merge option in panda dataframe 

```python
final_df = pd.merge(df, df_country,on = 'Country Code', how='left')
final_df.head(10)
```

```python
# to check data type we use 
final_df .dtypes
```

```python
# Here from value_counts we find how many record of any particular country has in data.
final_df.Country.value_counts()
```

```python
# this gives list of country to corresponding records
country_name = final_df.Country.value_counts().index
country_name
```

```python
#This gives values corresponding to country_name
country_records = final_df.Country.value_counts().values
country_records
```

 **Top three Countries that uses zomato** 

```python

plt.pie(country_records[:3], labels=country_name[:3],autopct='%1.2f%%' )
plt.show()
```

```python
final_df.columns
```

Observation : Zomato is used highest in India followed by United States and then United Kingdom.

 **Find the countries name that have given zero rating ?**

```python
# Method 1
final_df[final_df['Rating color']=='White'].groupby('Country').size().reset_index().rename(columns={0:'Not rated count'})
```

```python
# Method 2
final_df.groupby(['Aggregate rating','Country']).size().reset_index().head(5)
```

Observation:
1. Maximum non rating or 0 rating are from Indian Customers 

```python
Ratings = final_df.groupby(['Aggregate rating','Rating color', 'Rating text']).size().reset_index().rename(columns={0:"Rating Count"})
```

```python
Ratings
```

Observation 
1. Rating between 4.5 - 4.9 is Excellent 
2. Rating between 4.0 - 4.4 is Very Good 
3. Rating between 3.5 - 3.9 is  Good
4. Rating between 2.5 - 3.4 is Average
5. Rating between 1.8 - 2.4 is poor
6. Rating 0 is Not rated

```python
import matplotlib
matplotlib.rcParams['figure.figsize']= (12,6) # To increase the size of figure
sns.barplot(data= Ratings ,x='Aggregate rating',y= 'Rating Count', hue = "Rating color", palette= ['Blue', 'red','orange','yellow','green','green'])
```

Observation
1. Not Rated count is very high
2. Maximum number of rating is between 2.5 to 3.4
